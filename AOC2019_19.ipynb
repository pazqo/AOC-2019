# AOC2019 - Day 19

Today's challenge: [Day 19](https://adventofcode.com/2019/day/19)

Data are manually downloaded into `./data/input_19.txt`

```python
with open('./data/input_19.txt') as f:
    raw = f.read()
```

```python
from Intcode import Intcode

def check_value(x, y, data):
    program = Intcode(data.copy())
    program.input(x)
    program.input(y)
    return next(program)
```

```python
data = [int(x) for x in raw.strip().split(",")]
```

```python
field = []
for x in range(50):
    row = []
    for y in range(50):
        row.append(check_value(x, y, data.copy()))
    field.append(row)
```

```python
m = ['.','#',' ']
res = "\n".join(["".join([m[x] for x in f]) for f in field])
print(res)
```

```python
res.count('#')
```

```python
s = 100
x, y = s, s
while True:
    found_1 = False
    while not found_1:
        # find upper-right corner from diagonal (it's all below y=x)
        y += 1
        found_1 = check_value(x, y, data)
    x_, y_ = x - (s-1), y + (s-1) # this is the other corner
    if check_value(x_, y_, data.copy()) == 1:
        print("corner", x, y)
        print("other corner", x_, y_)
        print("answer: ", 10000*x_ + y)
        break
    else:
        # move to next column
        x += 1
        y = x
```
