# AOC2019 - Day 17

Today's challenge: [Day 17](https://adventofcode.com/2019/day/17)

Data are manually downloaded into `./data/input_17.txt`

```python
with open('./data/input_17.txt') as f:
    raw = f.read()
```

```python
from Intcode import Intcode
```

```python
from collections import defaultdict
data = [int(x) for x in raw.strip().split(",")]
data_ = data[:]
```

```python
def nbs(t):
    x, y = t
    return [(x-1,y),(x+1,y),(x,y-1),(x,y+1)]
```

```python
p = Intcode(data_)
p.run()
```

```python
rows = "".join([chr(x) for x in p.res]).split("\n")
path = set()
for y, row in enumerate(rows):
    for x, c in enumerate(row):
        if c == '#':
            path.add((x,y))
```

```python
params = []
for t in path:
    if all(t_ in path for t_ in nbs(t)):
        params.append(t)
sum([x*y for x, y in params])
```

```python
# Now part 2
data_ = data[:]
data_[0] = 2
p = Intcode(data_)
```

```python
sol = """A,A,B,C,A,C,B,C,A,B
L,4,L,10,L,6
L,6,L,4,R,8,R,8
L,6,R,8,L,10,L,8,L,8
n
"""
for c in sol:
    p.input(ord(c))
p.run()
```
